## 1.1 Business Context
Enterprises face an ever-growing threat of network intrusions — unauthorised access attempts, denial-of-service floods, and privilege escalation exploits. A single undetected breach costs on average **USD 4.45M** (IBM Cost of a Data Breach Report, 2023). Traditional signature-based Intrusion Detection Systems (IDS) fail to detect zero-day attacks and generate excessive false positives that overwhelm Security Operations Centre (SOC) teams.

**Machine learning–based anomaly detection** can:
- Catch novel attack patterns not seen in signatures
- Reduce false-positive fatigue by learning normal traffic baselines
- Prioritise alerts by attack likelihood score (probability output)

## 1.2 Problem Statement
> *Given a network connection record with 41 features (duration, protocol, byte counts, error rates, etc.), determine whether it represents **normal** traffic or an **attack** (anomaly).*

## 1.3 Task Type
| Dimension | Decision |
|---|---|
| Primary task | **Binary Classification** (Normal vs. Attack) |
| Secondary task | **Multi-class classification** (DoS / Probe / R2L / U2R) |
| Unsupervised component | **Anomaly detection** (Isolation Forest + Autoencoder) for label-free benchmarking |

## 1.4 Success Metrics
### Technical KPIs
| Metric | Target | Rationale |
|---|---|---|
| **F1-Score** (primary) | ≥ 0.97 | Balances precision + recall; critical when attack data is imbalanced |
| **AUC-ROC** | ≥ 0.99 | Measures ranking quality across all thresholds |
| **False Positive Rate (FPR)** | ≤ 0.02 | SOC team can only investigate ≤2% false alarms cost-effectively |
| **Recall (Sensitivity)** | ≥ 0.98 | Missing an attack (FN) is more costly than a false alarm |

### Business KPIs
| KPI | Estimate |
|---|---|
| Reduction in undetected breaches | -80% vs. signature IDS |
| SOC analyst alert fatigue reduction | -40% (fewer FP alerts) |
| Mean Time to Detect (MTTD) | < 60 seconds (real-time scoring) |
| Annual cost savings (avoided breach) | ~USD 3.5M per prevented incident |

## 1.5 Dataset Choice & Justification
**Dataset:** NSL-KDD (University of New Brunswick, 2009)  
**Why NSL-KDD?**
- Removes duplicate records from KDD Cup 1999 (which caused 78% of data to be duplicates — skewing results)
- ~125,973 training / 22,544 test records with 41 features
- 4 attack categories: DoS, Probe, R2L, U2R
- Industry-standard benchmark enabling fair comparison with published literature
